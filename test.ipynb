```python
import subprocess

def get_secret(project_id: str, secret_id: str, version_id: str = "latest") -> str:
    """
    Get secret from GCP Secret Manager using gcloud CLI.
    Uses the active gcloud account (patrick@sematryx.com).
    """
    result = subprocess.run(
        [
            "gcloud", "secrets", "versions", "access", version_id,
            "--secret", secret_id,
            "--project", project_id
        ],
        capture_output=True,
        text=True,
        check=True
    )
    return result.stdout.strip()

# Usage
api_key = get_secret("sematryx-481510", "sematryx-test-api-key")
```

```python
from sematryx import Sematryx

# Initialize the client
client = Sematryx(api_key=api_key)

# Define your objective function
def sphere(x):
    return sum(xi**2 for xi in x)

# Now try the actual optimization with the function
# The SDK should automatically convert it to a string expression
print("Running optimization with Python function...")
result = client.optimize(
    objective="minimize",
    variables=[
        {"name": "x", "bounds": (-5, 5)},
        {"name": "y", "bounds": (-5, 5)},
    ],
    objective_function=sphere,
)

print(f"Solution: {result.solution}")
print(f"Value: {result.value}")
```
